```python
# Importing important libraries
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image, ImageOps
```

```python
# To open an image in python, open() function in Image module of PIL library is useful.
# Similarly, matplotlib is used for reading and writing images.
# matplotlib uses 2 methods - imread(): to read images, imshow(): to display images.
# PIL - Python Imaging Library.
```

```python
# Please give appropriate file path for your own image for analysis.
```

```python
# Importing the image data
img = np.array(Image.open('Tiger.jpeg'))
plt.figure(figsize = (8, 8))
plt.imshow(img)
```

```python
# Getting the details of image
print('Number of dimensions : ', img.ndim)
print('Image shape : ', img.shape)
print('Dtype : ', img.dtype)
print(img[20, 20]) # This will give pixel values at RGB
print(img[:, :, 2].min())
```

```python
# Saving an image in n-dimensional array
path = 'Tiger_1.jpeg'
pil_img = Image.fromarray(img)
pil_img.save(path)
```

```python
# Getting 1-D View
img = Image.open('Tiger.jpeg')
arr = np.array(img)

# Recording shape of original arr
shape = arr.shape
print(shape)

# Making it 1 dimensional
flat_arr = arr.ravel()
print(flat_arr)

# Converting it back to a matrix
vector = np.matrix(flat_arr)
print(vector)
```

```python
# Making some changes 
vector[:,::10] = 128

# Reform a numpy array of original shape
arr2 = np.asarray(vector).reshape(shape)

# Making a PIL Image
img2 = Image.fromarray(arr2, 'RGB')
img2.show()
```

```python
# Histogram for pixel densities
img = np.array(Image.open('Tiger.jpeg'))
img_flat = img.flatten()
plt.hist(img_flat, bins = 200, range = [0, 256])
plt.title('Number of pixels in each intensity value')
plt.xlabel('Intensity')
plt.ylabel('Number of Pixels')
plt.show()
```

```python
# The above graph will give you idea about the quality of image.
```

```python
# Flipping an image
fig = plt.figure(figsize = (10, 10))
fig.add_subplot(1,2,1)
plt.imshow(np.flipud(img))
fig.add_subplot(1,2,2)
plt.imshow(np.fliplr(img))
```

```python
# Trimming/Cropping an image
fig = plt.figure(figsize = (10, 10))
fig.add_subplot(1,2,1)
plt.imshow(img)
plt.title('Original')
img0 = img[128:-128, 128:-128,:]
fig.add_subplot(1,2,2)
plt.imshow(img0)
plt.title('Trimmed')
```

```python
# Padding Black Spaces
img = np.array(Image.open('Tiger.jpeg'))
img_grey = img.sum(2) / (255*3)
img0 = img_grey.copy()
img0 = np.pad(img0, ((100,100), (100,100)), mode = 'constant')
plt.imshow(img0)
```

```python
# Negative image
fig = plt.figure(figsize = (10, 10))
img_grey = (255*3) - img_grey
fig.add_subplot(1,2,1)
plt.imshow(img_grey)
plt.title('Negative of Grey Image')
fig.add_subplot(1,2,2)
plt.imshow(img)
plt.title('Actual Image')
```
